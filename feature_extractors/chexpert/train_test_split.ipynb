```python
import pandas as pd
df = pd.read_csv('../../MIMIC-CXR-JPG/mimic-cxr-jpg/2.1.0/dicom-to-additional.csv')
df
```

```python
from sklearn.model_selection import train_test_split
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["respirator"],
    random_state=42
)
```

```python
train_df["split"] = "train"
val_df["split"] = "test"

df_split = pd.concat([train_df, val_df])
df_split.to_csv("annotations_with_split.csv", index=False)
```
